```python
from sklearn import datasets
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
X, y = datasets.make_regression(n_samples=250, n_features=2, noise=20, random_state=0, bias=50)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
import numpy as np
o_train = np.ones([X_train.shape[0],1], X_train.dtype )
o_test = np.ones([X_test.shape[0], 1], X_test.dtype)
X_train = np.concatenate((o_train, X_train), axis=1)
X_test = np.concatenate((o_test, X_test), axis=1)
```

```python
plt.scatter(X_test[:,1], y_test, c='red', edgecolors='white')
plt.scatter(X_train[:, 1], y_train, c='orange', edgecolors='white')
plt.show()
```

```python
theta = np.zeros([3, 1], dtype=np.float32)
```

```python
print(theta)
```

```python
theta = np.matmul(np.linalg.inv(np.matmul(X_train.T, X_train)), np.matmul(X_train.T, y_train))
theta = np.reshape(theta, [3, 1])
```

```python
p_y_test = np.matmul(X_test, theta).flatten()
```

```python
plt.figure(1)
plt.scatter(y_test, p_y_test, c='red', edgecolors='white')
plt.xlabel('Actual values')
plt.ylabel('Predicted values')
plt.title('Test set: Actual vs Predicted')
plt.show()
```

```python
p_y_train = np.matmul(X_train, theta).flatten()

plt.figure(0)
plt.scatter(y_train, p_y_train, c='purple', edgecolors='white')
plt.xlabel('Actual values')
plt.ylabel('Predicted values')
plt.title('Training set: Actual vs Predicted')
plt.show()
```
